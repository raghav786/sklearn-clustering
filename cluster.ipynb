```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
import numpy as np
import random
```

# dataset

```python
samples=pd.read_csv('https://gist.githubusercontent.com/curran/a08a1080b88344b0c8a7/raw/d546eaee765268bf2f487608c537c05e22e4b221/iris.csv',header=0,usecols=[0,1,2,3])
samples=samples.values
print(samples[:10])
```

# Calculating inertia

```python
inertia=[]
for i in range(1,21):
    model=KMeans(n_clusters=i)
    model.fit(samples)
    inertia.append(model.inertia_)
print(inertia,len(inertia))
```

```python
plt.plot(range(1,21),inertia)
plt.xlabel('clusters')
plt.ylabel('inertia')
plt.show()
```

# Create new samples using numpy random

```python
new_samples=np.random.rand(150,4)
new_samples=new_samples*10
print(new_samples[:10])
```

```python
model=KMeans(n_clusters=3)
model.fit(samples)
labels=model.predict(new_samples)
print(labels)
```

# Plotting Cluster Centers

```python
x=samples[:,0]#sepal length
y=samples[:,2]#petal length
centroids=model.cluster_centers_
plt.scatter(x,y,c=labels)
plt.scatter(centroids[:,0],centroids[:,2],marker='D',s=150,color='black')
plt.show()
```
